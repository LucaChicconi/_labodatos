(d) (Difícil) Considerando las veces que salieron 3 caras seguidas, la proporción de veces que en la siguiente tirada sale nuevamente cara.

````python
# Y ésto es lo que tiene para opinar respecto al (d)
def prop_cara_tras_tres_seguidas(n):
    """
    Dada la secuencia de n tiradas, busca los momentos donde
    hubo exactamente 3 caras consecutivas y mira la siguiente tirada.
    """
    s = np.random.choice([0, 1], size=n)  # 1=cara, 0=ceca
    
    total = 0
    cara_siguiente = 0
    
    for i in range(3, n):
        if s[i-3] == 1 and s[i-2] == 1 and s[i-1] == 1:
            total += 1
            if s[i] == 1:
                cara_siguiente += 1
    
    if total == 0:
        return None  # no hubo tres seguidas
    return cara_siguiente / total

Ns = [1000, 10000, 100000, 1000000]
for n in Ns:
    prop = prop_cara_tras_tres_seguidas(n)
    print(f"n={n:>8}: P(cara | 3 caras seguidas) = {prop:.6f}")

# La moneda NO tiene memoria → el límite es 0.5
````
